```python
# Data Handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries Imported Successfully")
```

```python
# Load Dataset
df = pd.read_csv(r"C:\Users\senan\Downloads\credit_card_dataset\creditcard.csv")

# First 5 rows
df.head()
```

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Column names
print("\nColumns:\n", df.columns)

# Dataset info
df.info()

# Statistical summary
df.describe()
```

```python
# Missing values
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

```python
# Duplicate rows
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Remove duplicate rows

df = df.drop_duplicates()

# Check new shape
print("New Shape:", df.shape)

# Verify duplicates removed
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.to_csv("dataset/credit_risk_cleaned.csv", index=False)
```

```python
# Fraud vs Normal transactions

df['Class'].value_counts()
```

```python
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal Transactions")
plt.show()
```

```python
labels = ["Normal", "Fraud"]

values = df['Class'].value_counts()

plt.figure(figsize=(6,4))

sns.barplot(x=labels, y=values.values)

plt.title("Transaction Distribution")

plt.ylabel("Count")

plt.show()
```

```python
fraud_percentage = (df['Class'].sum() / len(df)) * 100

print(f"Fraud Percentage: {fraud_percentage:.4f}%")
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='Class', y='Amount', data=df)

plt.title("Fraud vs Transaction Amount")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Time'], bins=50)

plt.title("Transaction Time Distribution")

plt.show()
```

```python
plt.figure(figsize=(20,15))

sns.heatmap(df.corr(), cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

Key Findings
1. Dataset contains highly imbalanced fraud distribution.
2. Fraudulent transactions represent a very small percentage of total transactions.
3. Transaction amount distribution is highly skewed.
4. Some PCA-transformed variables show correlation with fraud class.
5. Fraud detection requires imbalance handling techniques like SMOTE.
